```python
#Create an Interactive Calendar UI
import ipywidgets as widgets
from IPython.display import display
```

```python
date_dropdown = widgets.Dropdown(
    options=[str(d) for d in sorted(schedule.keys())],
    description="Date:",
    style={'description_width':'initial'}
)
```

```python
def show_event(date):
    date = int(date)
    event = schedule.get(date, ["No events scheduled."])
    print(f"üìÖ Events on {date} February:")
    for e in event:
        print(f"    -{e}")
```

```python
widgets.interactive(show_event, date=date_dropdown)
```
