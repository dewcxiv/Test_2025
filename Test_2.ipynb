```python
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
# List of example sentences
sentences = [
    "I love programming with Python!",
    "I hate bugs in my code.",
    "This project is very interesting.",
    "I'm feeling a little frustrated with the errors.",
    "Today is a great day to code!"
]
```

```python
def analyze_sentiment(sentence):
    blob = TextBlob(sentence)
    return blob.sentiment.polarity
```

```python
sentiment_scores = [analyze_sentiment(sentence) for sentence in sentences]
```

```python
plt.figure(figsize=(10, 5))
plt.bar(sentences, sentiment_scores, color=['green' if score > 0 else 'red' for score in sentiment_scores])
plt.xlabel("Sentences")
plt.ylabel("Sentiment Polarity")
plt.title("Sentiment Analysis of Example Sentences")
plt.xticks(rotation=45, ha='right')
plt.show()
```
