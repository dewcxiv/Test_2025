# Predict the number of concert DAY6 could have held in 2020 if COVID-19 hadn't happened

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Create a dataset
data = {
    'Year': [2015, 2017, 2018, 2019], # past years with tours
    'Concerts': [5, 25, 26, 31] # number of concerts held
}
df = pd.DataFrame(data)
```

```python
# Split data into features (X) and target (y)
X = df[['Year']]  # Feature (Year)
y = df['Concerts']  # Target (Number of Concerts)
```

```python
# Train a Linear Regression Model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict concerts for 2020
year_2020 = pd.DataFrame([[2020]], columns=['Year'])
predicted_concerts = model.predict(year_2020)
print(f"Predicted Number of DAY6 Concerts in 2020: {predicted_concerts[0]:.0f}")
```

```python
# Visualize the data and prediction
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X, model.predict(X), color='green', linestyle='dashed', label='Trend Line')
plt.scatter(2020, predicted_concerts, color='blue', marker='x', s=200, label='Prediction for 2020')
plt.xlabel('Year')
plt.ylabel('Number of Concerts')
plt.title('DAY6 World Tour Concert Prediction')
plt.legend()
plt.show()
```


# Predict how many concerts they might have in 2025!


```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Updated dataset with more years (2015-2023)
data = {
    'Year': [2015, 2017, 2018, 2019, 2020, 2023],
    'Concerts': [5, 25, 26, 31, 39, 20]
}

df = pd.DataFrame(data)

# Step 3: Train the Linear Regression Model
X = df[['Year']]  # Feature (Year)
y = df['Concerts']  # Target (Number of Concerts)
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict concerts for 2025
year_2025 = pd.DataFrame([[2025]], columns=['Year'])
predicted_concerts = model.predict(year_2025)
print(f"Predicted Number of DAY6 Concerts in 2025: {predicted_concerts[0]:.0f}")

# Step 5: Visualize the Data & Prediction
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='green', linestyle='dashed', label='Trend Line')
plt.scatter(2025, predicted_concerts, color='red', marker='x', s=200, label='Prediction for 2025')
plt.xlabel('Year')
plt.ylabel('Number of Concerts')
plt.title('DAY6 World Tour Concert Prediction (2025)')
plt.legend()
plt.show()
```

This means DAY6 might hold 36 concerts worldwide in 2025 if trends continue!🍀
